
# 🧠 Brain Rot Analytics — Model Training Notebook

**المشروع:** التنبؤ بمرحلة التشتت الرقمي (`Brainrot_Stage`) لدى الطلاب المصريين
باستخدام 7 عوامل سلوكية يومية فقط (بدون Data Leakage)، ومقارنة 3 موديلات
Machine Learning لاختيار الأفضل بناءً على macro F1-score.

**الفئات المستهدفة (4):** `Healthy` (61%) / `Critical` (18%) / `Advanced` (12.5%) / `Casual` (8.5%)

هذا الـ Notebook هو النسخة الرسمية لتدريب النموذج، ومُصمَّم ليُعرض مباشرة في مناقشة التخرج
بكل الرسومات والنتائج ظاهرة بدون الحاجة لإعادة تشغيله.


## 1) Imports | استدعاء المكتبات

```python

import os
import warnings
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, f1_score, accuracy_score
from xgboost import XGBClassifier

from utils import (
    load_raw_data,
    clean_data,
    prepare_features,
    FEATURE_COLUMNS,
    TARGET_COLUMN,
    STAGE_ORDER,
)

warnings.filterwarnings("ignore")
sns.set_style("darkgrid")
plt.rcParams["figure.facecolor"] = "white"

RANDOM_STATE = 42
DATA_PATH = "data/BrainRot_Final_Dataset.csv"
MODEL_DIR = "model"
os.makedirs(MODEL_DIR, exist_ok=True)

print("Libraries loaded successfully. | تم تحميل المكتبات بنجاح")
```

## 2) Load Data | تحميل البيانات

نقرأ ملف `BrainRot_Final_Dataset.csv` (5000 صف، 30 عمود) ونلقي نظرة أولى عليه.

```python

df_raw = load_raw_data(DATA_PATH)
print("Shape:", df_raw.shape)
df_raw.head()
```

```python

df_raw.info()
```

## 3) Exploratory Data Analysis (EDA) | استكشاف البيانات

قبل أي تنظيف، نفحص:
- توزيع الفئات المستهدفة (`Brainrot_Stage`) — للتأكد من عدم توازن الفئات (Class Imbalance).
- عدد القيم الفاضية (Missing Values) في كل عمود.
- إحصائيات وصفية عامة للأعمدة الرقمية.


```python

# Target class distribution | توزيع الفئات المستهدفة
target_counts = df_raw[TARGET_COLUMN].value_counts()
target_pct = (target_counts / len(df_raw) * 100).round(1)
print(target_counts)
print("\nPercentages:\n", target_pct)

fig, ax = plt.subplots(figsize=(7, 5))
colors = {"Healthy": "#22C55E", "Casual": "#EAB308", "Advanced": "#F97316", "Critical": "#F43F5E"}
bar_colors = [colors[s] for s in target_counts.index]
ax.bar(target_counts.index, target_counts.values, color=bar_colors)
ax.set_title("Brainrot_Stage Distribution (Class Imbalance)")
ax.set_ylabel("Count")
for i, v in enumerate(target_counts.values):
    ax.text(i, v + 30, str(v), ha="center", fontweight="bold")
plt.tight_layout()
plt.show()
```

```python

# Missing values | القيم الفاضية
missing = df_raw.isnull().sum()
missing = missing[missing > 0]
print("Columns with missing values:")
print(missing)
```

```python

# Descriptive statistics for the key numeric columns we care about
key_numeric_cols = ["Age", "Total_Reels_Watched", "Coffee_Consumed_Per_Day",
                     "Focus_Sessions_Count", "Study_Hours"]
df_raw[key_numeric_cols].describe().T
```

```python

fig, axes = plt.subplots(1, len(key_numeric_cols), figsize=(20, 4))
for ax, col in zip(axes, key_numeric_cols):
    sns.histplot(df_raw[col].dropna(), kde=True, ax=ax, color="#7C5CFC")
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

## 4) Data Cleaning | تنظيف البيانات

الأعمدة `Age, Age_Group, Region, Device_Type, Is_Smoker, Base_Focus_Level, Username`
فيها 51 صف فاضي من أصل 5000. **القرار:** عمل impute (median/mode) بدل drop، عشان نحافظ
على حجم العينة بالكامل. الدالة `clean_data()` في `utils.py` بتطبق نفس المنطق ده بالظبط
(ونفس النسخة بيستخدمها `app.py` وقت التنبؤ، عشان التناسق بين التدريب والتطبيق).


```python

df = clean_data(df_raw)
print("Shape after cleaning:", df.shape)
print("Remaining missing values:", df.isnull().sum().sum())
df[["Age", "Device_Type"]].isnull().sum()
```

## 5) Feature Selection & Encoding | اختيار وترميز الفيتشرز

نستخدم الـ **7 مدخلات المسموح بيها فقط**:
`Age, Total_Reels_Watched, Coffee_Consumed_Per_Day, Focus_Sessions_Count, Study_Hours, Is_Late_Night, Device_Type`

⚠️ **تنبيه Data Leakage:** تم استبعاد `Brainrot_Exposure_Score`, `Wellbeing_Score`,
`Attention_Span_Level`, `Aura_Color_Code`, `Coffee_Level`, `Smoking_Status` من التدريب لأنها
مشتقة أو مرتبطة مباشرة بالـ target.

نعمل One-Hot Encoding لعمود `Device_Type` فقط (`Is_Late_Night` أصلاً 0/1).


```python

X = prepare_features(df)  # selects the 7 features + one-hot encodes Device_Type
y_labels = df[TARGET_COLUMN].astype(pd.CategoricalDtype(categories=STAGE_ORDER))

trained_columns = X.columns.tolist()
print("Final training columns:", trained_columns)

label_to_int = {label: i for i, label in enumerate(STAGE_ORDER)}
int_to_label = {i: label for label, i in label_to_int.items()}
y = y_labels.map(label_to_int).astype(int)

X.head()
```

## 6) Train/Test Split | تقسيم البيانات

تقسيم Stratified 80/20 عشان نحافظ على نفس نسب الفئات في التدريب والاختبار.

```python

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Train shape:", X_train.shape, " Test shape:", X_test.shape)
```

## 7) Model 1 — Logistic Regression (Baseline) | نموذج أساسي

نبدأ بـ Logistic Regression كـ **baseline** بسيط وسريع، مع `class_weight='balanced'`
لمعالجة عدم توازن الفئات. ده بيدينا نقطة مرجعية نقارن بيها الموديلات الأعقد.


```python

results = {}

log_reg = LogisticRegression(class_weight="balanced", max_iter=1000, random_state=RANDOM_STATE)
log_reg.fit(X_train_scaled, y_train)

pred_lr = log_reg.predict(X_test_scaled)
f1_lr = f1_score(y_test, pred_lr, average="macro")
acc_lr = accuracy_score(y_test, pred_lr)

results["LogisticRegression"] = {"model": log_reg, "f1_macro": f1_lr, "accuracy": acc_lr, "uses_scaled": True}
print(f"LogisticRegression -> macro F1: {f1_lr:.4f} | accuracy: {acc_lr:.4f}")
```

## 8) Model 2 — Random Forest | غابة عشوائية

نجرب Random Forest لأنه قوي مع البيانات الجدولية (tabular)، بيتعامل كويس مع العلاقات
غير الخطية، وأقل حساسية لمشكلة الـ scaling. نستخدم `RandomizedSearchCV` لضبط أهم
الـ hyperparameters (عدد الأشجار، العمق، الحد الأدنى للتقسيم/الأوراق).


```python

rf_param_dist = {
    "n_estimators": [200, 300, 400],
    "max_depth": [None, 8, 12, 16],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
}
rf_search = RandomizedSearchCV(
    RandomForestClassifier(class_weight="balanced", random_state=RANDOM_STATE),
    rf_param_dist,
    n_iter=10,
    scoring="f1_macro",
    cv=3,
    random_state=RANDOM_STATE,
    n_jobs=-1,
)
rf_search.fit(X_train, y_train)
rf_best = rf_search.best_estimator_

pred_rf = rf_best.predict(X_test)
f1_rf = f1_score(y_test, pred_rf, average="macro")
acc_rf = accuracy_score(y_test, pred_rf)

results["RandomForest"] = {"model": rf_best, "f1_macro": f1_rf, "accuracy": acc_rf, "uses_scaled": False}
print(f"RandomForest -> macro F1: {f1_rf:.4f} | accuracy: {acc_rf:.4f}")
print("Best params:", rf_search.best_params_)
```

## 9) Model 3 — XGBoost | تعزيز التدرج

نجرب XGBoost لأنه من أقوى الخوارزميات على البيانات الجدولية وعادة بيدي أداء ممتاز.
بما إن XGBoost مفيهوش `class_weight='balanced'` مباشرة، بنحسب `sample_weight` يدويًا
لكل صف حسب ندرة فئته، ونستخدم `RandomizedSearchCV` كمان لضبط الـ hyperparameters.


```python

class_counts = y_train.value_counts()
n_samples = len(y_train)
n_classes = len(class_counts)
class_weights = {c: n_samples / (n_classes * cnt) for c, cnt in class_counts.items()}
sample_weights = y_train.map(class_weights).values

xgb_param_dist = {
    "n_estimators": [200, 300, 400],
    "max_depth": [3, 5, 7],
    "learning_rate": [0.03, 0.05, 0.1],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0],
}
xgb_base = XGBClassifier(
    objective="multi:softprob",
    num_class=len(STAGE_ORDER),
    eval_metric="mlogloss",
    random_state=RANDOM_STATE,
    n_jobs=-1,
)
xgb_search = RandomizedSearchCV(
    xgb_base, xgb_param_dist, n_iter=10, scoring="f1_macro",
    cv=3, random_state=RANDOM_STATE, n_jobs=-1,
)
xgb_search.fit(X_train, y_train, sample_weight=sample_weights)
xgb_best = xgb_search.best_estimator_

pred_xgb = xgb_best.predict(X_test)
f1_xgb = f1_score(y_test, pred_xgb, average="macro")
acc_xgb = accuracy_score(y_test, pred_xgb)

results["XGBoost"] = {"model": xgb_best, "f1_macro": f1_xgb, "accuracy": acc_xgb, "uses_scaled": False}
print(f"XGBoost -> macro F1: {f1_xgb:.4f} | accuracy: {acc_xgb:.4f}")
print("Best params:", xgb_search.best_params_)
```

## 10) Model Comparison & Selection | مقارنة النتائج واختيار الأفضل

نقارن الموديلات الثلاثة بناءً على **macro F1-score** (مش accuracy بس)، لأن الداتا
غير متوازنة والـ macro F1 بيدي وزن متساوٍ لكل فئة حتى لو كانت قليلة العدد.


```python

comparison_df = pd.DataFrame({
    name: {"macro_F1": r["f1_macro"], "accuracy": r["accuracy"]}
    for name, r in results.items()
}).T.sort_values("macro_F1", ascending=False)

comparison_df
```

```python

best_name = comparison_df.index[0]
best_model = results[best_name]["model"]
best_uses_scaled = results[best_name]["uses_scaled"]

print(f"🏆 BEST MODEL: {best_name}  (macro F1 = {results[best_name]['f1_macro']:.4f}, "
      f"accuracy = {results[best_name]['accuracy']:.4f})")

X_test_for_report = X_test_scaled if best_uses_scaled else X_test
y_pred_best = best_model.predict(X_test_for_report)

report = classification_report(y_test, y_pred_best, target_names=STAGE_ORDER, digits=3)
print(report)

with open(os.path.join(MODEL_DIR, "classification_report.txt"), "w") as f:
    f.write(f"Best model: {best_name}\n\n")
    f.write(report)
```

## 11) Confusion Matrix | مصفوفة الالتباس

توضح أداء الموديل الفائز على كل فئة على حدة.

```python

cm = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Purples",
            xticklabels=STAGE_ORDER, yticklabels=STAGE_ORDER)
plt.title(f"Confusion Matrix - {best_name}")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.savefig(os.path.join(MODEL_DIR, "confusion_matrix.png"), dpi=150)
plt.show()
```

## 12) Feature Importance | أهمية الفيتشرز

بتوضح إن `Total_Reels_Watched` هو أقوى عامل مؤثر في التنبؤ، يليه `Focus_Sessions_Count` و `Study_Hours`.

```python

if hasattr(best_model, "feature_importances_"):
    importances = best_model.feature_importances_
elif hasattr(best_model, "coef_"):
    importances = np.mean(np.abs(best_model.coef_), axis=0)
else:
    importances = np.zeros(len(trained_columns))

fi_df = pd.DataFrame({"feature": trained_columns, "importance": importances}) \
          .sort_values("importance", ascending=False)
fi_df.to_csv(os.path.join(MODEL_DIR, "feature_importance.csv"), index=False)

plt.figure(figsize=(7, 5))
sns.barplot(data=fi_df, x="importance", y="feature", palette="viridis")
plt.title(f"Feature Importance - {best_name}")
plt.tight_layout()
plt.savefig(os.path.join(MODEL_DIR, "feature_importance.png"), dpi=150)
plt.show()

fi_df
```

## 13) Save Final Model & Artifacts | حفظ الموديل النهائي

نحفظ الموديل الفائز والـ scaler والـ metadata بنفس الأسماء والمسارات المستخدمة في
`app.py` (`model/brainrot_model.pkl`, `model/scaler.pkl`, `model/metadata.pkl`) عشان
التطبيق يفضل شغال زي ما هو من غير أي تعديل.


```python

joblib.dump(best_model, os.path.join(MODEL_DIR, "brainrot_model.pkl"))
joblib.dump(scaler, os.path.join(MODEL_DIR, "scaler.pkl"))
# Also saved as encoder.pkl for compatibility with the project spec / README file structure
joblib.dump(scaler, os.path.join(MODEL_DIR, "encoder.pkl"))

joblib.dump(
    {
        "trained_columns": trained_columns,
        "stage_order": STAGE_ORDER,
        "label_to_int": label_to_int,
        "int_to_label": int_to_label,
        "best_model_name": best_name,
        "uses_scaled": best_uses_scaled,
        "all_results": {k: v["f1_macro"] for k, v in results.items()},
    },
    os.path.join(MODEL_DIR, "metadata.pkl"),
)

print("✅ Saved model/brainrot_model.pkl, model/scaler.pkl, model/encoder.pkl, model/metadata.pkl")
print(f"✅ Final model: {best_name} | macro F1 = {results[best_name]['f1_macro']:.4f} | "
      f"accuracy = {results[best_name]['accuracy']:.4f}")
```
